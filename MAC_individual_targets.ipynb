```python
# Function to count author+year combinations in a string
def count_authors_year_combinations(s):
    return len(s.split(' ; '))

# Create a new column to store the count of author+year combinations
df['author_year_count'] = df['Paper(s)'].apply(count_authors_year_combinations)

# Identify rows where the count of author+year combinations is more than 4
df['more_than_4'] = df['author_year_count'] > 4

df
```
