```python
import pandas as pd
import numpy as np

pd.options.display.max_colwidth = 200
```

```python
MAC = pd.read_csv('MAC_DR0p8_formatted.csv', sep=',')
```

```python
# Within the 'parsed selection method' column, I have stored dictionaries that contain key:value entries. These \
# key:value entries encode the selection strategies used throughout the literature.

# At the present time, these are the available wavebands (or techniques.. some add'l formatting is still \
# necessary):
# Optical, Near-IR, Mid-IR, Infrared, Gamma-Ray, X-ray, BAT, Radio, UV, Modeling
```

```python
import ast

# this nested function allows you to input a waveband and technique that you're interested \
# and returns a function that you can use to filter the data based. This is useful for parsing \
# on selection technique
def get_filter_function(waveband, technique):
    def filter_function(row):
        # Access the value in the 'Parsed Selection Method' column
        selection_method = row['Parsed Selection Method']
        
        # If the value is a string, convert it back to a dictionary
        if isinstance(selection_method, str):
            try:
                selection_dict = ast.literal_eval(selection_method)
            except ValueError:
                return False
        else:
            selection_dict = selection_method

        # Check if the waveband is a key and the technique is one of its values
        return str(waveband) in selection_dict and str(technique) in selection_dict[str(waveband)]

    return filter_function

```

```python
# Here are some commands that allow you to parse the table based on selection methodology

# If you want to only include objects selected using X-rays in some what, use this:
filter_function = get_filter_function('X-ray', 'Imaging')
filtered_df = MAC[MAC.apply(filter_function, axis=1)]

# this grabs any objects with a key value of 'X-ray' in the Selection Method dictionary.

print(len(filtered_df)) # printing the length of the dataframe so we can see the number of objects
filtered_df  # and now we'll have the code show us the dataframe so we can visually inspect it
```

```python
# If you want to parse the table for objects that specifically used optical fiber spectroscopy at some point \
# in their selection method, use this:
filter_function = get_filter_function('Optical', 'Fiber Spectroscopy')
filtered_df = MAC[MAC.apply(filter_function, axis=1)]
# this accesses the 'Optical' key in the dictionary, and then retains only objects that have 'Fiber Spectroscopy' \
# as a value in the dictionary under the optical key.

filtered_df # and now we'll have the code show us the dataframe so we can visually inspect it

# be mindful: right now, the jupyter notebook is not set up to show you all of the rows; this cell provides only \
# a truncated view of the dataframe (but we can fix that later)
```

```python
# Here we have some commands to parse the table based on physical separation:
goodsep = MAC[(MAC['Sep(kpc)']<10) & (MAC['Sep(kpc)']>0)] 
# this takes only objects with separations less than 10 kpc but GREATER than 0 kpc (ensuring we don't look at \
# objects with a -99 flag for their separation)
# This cut will naturally grab dual AGNs and gravitationally-bound binary candidates

# and angular separation:
goodang = MAC[(MAC['Sep']<10) & (MAC['Sep']>0.1)]
# this takes only objects with separations less than 20'' but GREATER than 0.1'' (ensuring we don't look at \
# objects with a -99 flag for their separation)


# and even redshift (here we're cutting out high redshift and focusing on local redshift <0.1)\
goodz = MAC[(MAC['z1']<0.1) & (MAC['z1']!=-99)]
# here we've clipped on z1 and ensured that z1 does not have the -99 flag (which means the data is missing)
```

```python
# run this cell to print out the number of objects that we found in each of the previous three tables
print('Multi-AGNs and candidates with physical separations <10 kpc:'+str(len(goodsep)))
print('Multi-AGNs and candidates with angular separations <10 arcseconds:'+str(len(goodang)))
print('Multi-AGNs and candidates with redshifts <0.1:'+str(len(goodz)))
```
